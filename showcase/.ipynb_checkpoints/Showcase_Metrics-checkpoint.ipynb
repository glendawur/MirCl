### Basic principles of metrics subpackage

There are two types of functions for each index: \<*generic_name*\>() and \<*generic_name*\>\_matrix(). First one is computing the value of index for one partition and receives only one array of labels of size (N, ), while second type of functions is aimed to receive matrices with multiple partitions or with intermediate values for multiplt partitions.

### Analytical version of Elbow

```python
%time WSS = wss_matrix(X=x, L=L)
```

```python
elbow_values_small_step = elbow(WSS, (1, 1), aggregation=np.mean)
elbow_values_double_step = elbow(WSS, (2, 2), aggregation=np.mean)
```

```python
k_s, idx_s = find_optimal(elbow_values_small_step,
             method='elbow',
             k_range= np.arange(2, 21),
             to_plot=False)
k_d, idx_d = find_optimal(elbow_values_double_step,
             method='elbow',
             k_range= np.arange(2, 21),
             to_plot=False)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (20, 5))
ax[0].set_title('Conventional Intertia')
ax[0].plot(np.arange(2, 21), WSS.mean(axis = 1), c='black')
ax[0].vlines([k_s, k_d], color=['blue', 'red'], ymin = 0, ymax=np.max(WSS.mean(axis = 1)))

ax[0].plot([k_d-2, k_d, k_d+2],
           [WSS.mean(axis = 1)[idx_d-2], WSS.mean(axis = 1)[idx_d], WSS.mean(axis = 1)[idx_d+2]], c='red')
ax[0].plot([k_s-1, k_s, k_s+1],
           [WSS.mean(axis = 1)[idx_s-1], WSS.mean(axis = 1)[idx_s], WSS.mean(axis = 1)[idx_s+1]], c='blue')
ax[0].set_ylim([0, WSS.mean(axis=1).max()])
ax[0].set_xlim([2, 20])
ax[0].set_xticks(np.arange(2, 21))


ax[1].set_title('Regular (Conventional) Elbow')
ax[1].plot(np.arange(2, 21), elbow_values_small_step, c='blue')
ax[1].vlines([k_s], color='green', ymin=0, ymax=np.max(elbow_values_small_step[1:-1])+1)
ax[1].set_ylim([0, elbow_values_small_step[1:-1].max()+1])
ax[1].set_xlim([np.arange(2, 21)[0] + (k_s - idx_s), np.arange(2, 21)[-1] - (k_s - idx_s)])


ax[2].set_title('Two-step Elbow')
ax[2].plot(np.arange(2, 21), elbow_values_double_step, c='red')
ax[2].vlines([k_d], color='green', ymin=0,
            ymax=np.max(elbow_values_double_step[2:-2])+1)
ax[2].set_ylim([0, elbow_values_double_step[2:-2].max()+1])
ax[2].set_xlim([np.arange(2, 21)[0] + (k_d - idx_d), np.arange(2, 21)[-1] - (k_d - idx_d)])

for ax_i in ax:
    ax_i.set_xlabel('K = Number of clusters')
    ax_i.grid()
```

The *levels* parameter in Elbow function controls the window used to compute the ratio between two slopes in the given point. As of now, only fixed step-sizes are supported. One can do different step-size, for example:

```python
elbow_values_varying_step = elbow(WSS, (1, 3), aggregation=np.mean)
```

```python
example = 'pics\elbow_example.gif'

filenames = list()

for k, idx in zip(np.arange(2, 21)[1:-3], list(range(19))[1:-3]):
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    ax[0].plot(np.arange(2, 21), WSS.mean(axis = 1), color = 'black')
    ax[0].set_title(f'Values of WSS: {(WSS.mean(axis=1)[idx-1] - WSS.mean(axis=1)[idx]).round(3)} and {(WSS.mean(axis=1)[idx] - WSS.mean(axis=1)[idx+3]).round(3)}',
                   size = 'small')
    
    ax[1].plot(np.arange(2, 21), elbow_values_varying_step, color = 'black')
    ax[1].set_title(f'Value of Elbow: {elbow_values_varying_step[idx].round(3)}', size = 'small')
    
    ax[0].plot([k-1, k, k+3],
           [WSS.mean(axis = 1)[idx-1], WSS.mean(axis = 1)[idx], WSS.mean(axis = 1)[idx+3]], c='red')
    ax[0].vlines([k], color='red', ymin =0, ymax=np.max(WSS.mean(axis = 1)))
    ax[1].vlines([k], color='red', ymin=0, ymax=np.max(elbow_values_varying_step)+1)
    
    fig.suptitle(f'Number of clusters {k}')
    
    ax[0].set_xlim([2, 20])
    ax[0].set_xticks(np.arange(2, 21))
    ax[0].set_ylim([0, np.max(WSS.mean(axis = 1))])
    
    ax[1].set_xlim([3, 17])
    ax[1].set_xticks(np.arange(3, 18))
    ax[1].set_ylim([0, np.max(elbow_values_varying_step)+1])
    
    for ax_i in ax:
        ax_i.set_xlabel('K = Number of clusters', size = 'small')
        ax_i.grid()
    
    filename = f'elbow_{k}.png'
    filenames.append(filename)
    plt.savefig(filename, dpi=100)
    plt.close()
    
with imageio.get_writer(example, mode='I', duration=10/len(list(range(19))[1:-3])) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
        
for filename in set(filenames):
    os.remove(filename)
```

### Hartigan Rule

```python
%time hartigan_values = hartigan(X=x, L=L, SSW=WSS, aggregation=np.mean)
k, idx = find_optimal(hartigan_values,
                      method='hartigan',
                      k_range= np.arange(2, 21),
                      to_plot=False)
```

Being a rule of thumb, Hartigan Rule is not the best option to choose the number of clusters as it heavily relies on pre-defined heuristic that the optimal number of clusters if the smalles K with Hartigan Index less than 10

```python
plt.plot(np.arange(2, 21), hartigan_values)
plt.xlim([2, 20])
plt.ylim([0, np.max(hartigan_values[:-1])])
plt.xticks(np.arange(2,21))
plt.title(f'Hartigan Rule (Rule of Thumb), optimal K: {k}')
plt.hlines(y=10, xmin=2, xmax=20, color='red')
plt.grid()
plt.show()
```

### Calinski-Harabasz metric

```python
%time ch_values = calinski_harabasz_matrix(X=x, L=L, SSW=wss_matrix(X=x, L=L, method='conventional'),SSB=bss_matrix(X=x, L=L, method='conventional'),aggregation=np.mean)

k, idx = find_optimal(ch_values, method='calinski_harabasz', k_range= np.arange(2, 21), to_plot=True)
```

### Xu Index

```python
%time xu_values = xu_index_matrix(X=x, L=L,SSW=wss_matrix(X=x, L=L, method='conventional'), aggregation=np.mean)
k, idx = find_optimal(xu_values,
                      method='xu_index',
                      k_range= np.arange(2, 21),
                      to_plot=True)
```

### WB index

```python
%time wb_values = wb_index_matrix(L=L, SSW=wss_matrix(X=x, L=L, method='conventional'), SSB=bss_matrix(X=x, L=L, method='conventional'),aggregation=np.mean)
k, idx = find_optimal(wb_values,
                      method='wb_index',
                      k_range= np.arange(2, 21),
                      to_plot=True)
```

### Silhouette width

```python
%time sil_values = silhouette_matrix(X=x, L=L, aggregation=np.mean)
```

As you can see, silhouette is quite slow function as silhouette_width is computationally expensive and challenging. You can compute this matrix either using other implementation or use a shortcut:

```python
from mircl.metrics import silhouette_wss
```

```python
%time sil_values = silhouette_wss(X=x, L=L, SSW=wss_matrix(X=x, L=L, method='conventional'), aggregation = np.argmin)
```

```python
k, idx = find_optimal(sil_values,
                      method='silhouette',
                      k_range= np.arange(2, 21),
                      to_plot=True)
```
